# Transform/convert a columns by using logic function in a dataset

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    if name == "Kings XI Punjab":
        return "KXP"
    if name == "Delhi Daredevils":
        return "DD"
    else:
        return name
```

```python
df11=pd.read_csv('/kaggle/input/day-01/IPL Matches 2008-2020.csv',converters={'team1':rename})
df11
```

```python
import requests
import pandas as pd

df = pd.DataFrame()  # Initializing an empty DataFrame

for i in range(1, 429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)  # Concatenating the DataFrames

# Ensure to use the concatenated DataFrame 'df' for further processing
```
